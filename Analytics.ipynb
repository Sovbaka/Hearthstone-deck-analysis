### Давайте посмотрим на даты изменений баланса, а также масштабных обновлений, добавляющих новые карты.

Для справки, мы рассматриваем следующие события:
1. <span style="color:gray"> Патчи, в которых были изменения баланса, "Card changes"</span>
2. <span style="color:blue">Приключения, "Adventures", которые добавляют среднего размера сеты карт (их нужно получить, проходя приключения, что занимает какое-то время, можно попробовать увидеть небольшой лаг в переменных после добавления приключений)</span>
3. <span style="color:red">Expansions (по-русски их часто называют "патчи") – масштабные обновления, добавляющие новые карты и, начиная со второй, совпадающие с ротацией карт: самые старые уходят в вольный режим, остальные (вместе с только что добавленными) составляют карты стандартного режима</span>

```python
# Запишем даты приключений и патчей (можно и скачать, но мы, естественно, помним их наизусть)
adventures = {'Curse of Naxxramas': "22/07/2014", 'Blackrock Mountain': "02/04/2015", 'The League of Explorers': "12/11/2015",
              'One Night in Karazhan': "11/08/2016", "Galakrond's Awakening": '21/01/2020'}
for key in adventures:
    adventures[key] = pd.to_datetime(adventures[key], format = "%d/%m/%Y")

expansions = {'Goblins vs Gnomes': "08/12/2014", 'The Grand Tournament': "24/08/2015", 'Whispers of the Old Gods': "26/04/2016",
            'Mean Streets of Gadgetzan': '01/12/2016', "Journey to Un'Goro": '06/04/2017', 'Knights of the Frozen Throne': '10/08/2017',
            'Kobolds & Catacombs': '07/12/2017', 'The Witchwood': '12/04/2018', 'The Boomsday Project': '07/08/2018',
            "Rastakhan's Rumble": '04/12/2018'}  #есть еще, пока и так с избытком
for key in expansions:
    expansions[key] = pd.to_datetime(expansions[key], format = "%d/%m/%Y")
```

```python
last_date_observed = df.iloc[df.shape[0] - 1, 7]  # отметка, показывающая, до какого периода мы скачали аднные
```

```python
# Скачаем и запишем даты изменений баланса (Card changes)
card_changes_timestamps = []

URL = "https://hearthstone.fandom.com/wiki/Patches#List_of_patches"
page = requests.get(URL)

page_content = page.text
soup = BeautifulSoup(page_content, 'lxml')

patches = soup.find_all('tr')  #во всем документе ровно одна таблица, её table rows – патчи.

for patch in patches[1:]:
    try:
        table_data_list = patch.find('td', class_ = "text").find_all()
    
        if any([x.string == 'Card changes' for x in table_data_list]):
            card_changes_timestamps.append(pd.to_datetime(patch.find_all('td')[2].string))   #В таблице третьей колонкой идут даты
    except:
        continue
```

```python
plt.figure(figsize=(12, 5))
plt.plot(decks_per_day['creation_date'], decks_per_day['count'], marker='o', linestyle='', alpha = 0.7, markersize=5)
plt.xlabel('Дата')
plt.ylabel('Количество созданных за день колод')
plt.grid(axis='y')
plt.xticks(rotation=45)

for change in card_changes_timestamps:
    if change < last_date_observed:  # Выводим только то, что относится к данным, которые мы успели (sic!) скачать
        plt.axvline(change, color = 'grey', alpha = 0.3)

for adventure in adventures:
    if adventures[adventure] < last_date_observed:  
        plt.axvline(adventures[adventure], color = 'blue', alpha = 0.7)

for expansion in expansions:
    if expansions[expansion] < last_date_observed:  
        plt.axvline(expansions[expansion], color = 'red', alpha = 0.7)


plt.show()
```

Добавим в наш датафрейм немного данных о тех картах, которые содержатся в колодах

```python
#Подготовим сведения о том, какими карты являются по редкости, и какими способностями (keywords) они обладают
with open('allcards.json', mode = 'r', encoding = 'utf-8') as file:
    data = json.load(file)

card_names = [card['name'] for card in data] #Список со всеми картами
card_names = sorted(list(set(card_names))) 
card_names = [card for card in card_names if (('Rank 2' not in card) & ('Rank 3' not in card))] 

def get_card_rarity(card_name, cards_data):
    for card in cards_data:
        if card.get('name') == card_name:
            return card.get('rarity')
            
def get_card_mechanics(card_name, cards_data):
    for card in cards_data:
        if card.get('name') == card_name:
            return card.get('mechanics')

def get_card_referencedTags(card_name, cards_data):
    for card in cards_data:
        if card.get('name') == card_name:
            return card.get('referencedTags')

#Создадим соответствующие списки из карт, согласно их редкости и способностям

rare_cards = [card for card in card_names if get_card_rarity(card, data) == 'RARE']
epic_cards = [card for card in card_names if get_card_rarity(card, data) == 'EPIC']
legendary_cards = [card for card in card_names if get_card_rarity(card, data) == 'LEGENDARY']

battlecry_cards = [card for card in card_names
                    if get_card_mechanics(card, data) is not None
                    and 'BATTLECRY' in get_card_mechanics(card, data)]
deathrattle_cards = [card for card in card_names 
                     if get_card_mechanics(card, data) is not None 
                     and 'DEATHRATTLE' in get_card_mechanics(card, data)]
stealth_cards = [card for card in card_names 
                 if get_card_mechanics(card, data) is not None
                 and 'STEALTH' in get_card_mechanics(card, data)]
charge_cards1 = [card for card in card_names 
                if (get_card_mechanics(card, data) is not None)
                and 'CHARGE' in (get_card_mechanics(card, data))]
charge_cards2 = [card for card in card_names 
                if (get_card_referencedTags(card, data) is not None)
                and 'CHARGE' in (get_card_referencedTags(card, data))]

charge_cards = sorted(list(set(charge_cards1 + charge_cards2)))

discover_cards = [card for card in card_names 
                  if get_card_mechanics(card, data) is not None
                  and 'DISCOVER' in get_card_mechanics(card, data)]
rush_cards = [card for card in card_names 
              if get_card_mechanics(card, data) is not None
              and 'RUSH' in get_card_mechanics(card, data)]
taunt_cards = [card for card in card_names 
               if get_card_mechanics(card, data) is not None
               and 'TAUNT' in get_card_mechanics(card, data)]
```

Добавим колонки с информацией о картах:

```python
properties_rarity = {'rare_cards' : rare_cards, 'epic_cards': epic_cards, 'legendary_cards': legendary_cards}

properties_type = {'battlecry_cards': battlecry_cards,
             'deathrattle_cards': deathrattle_cards, 'stealth_cards': stealth_cards, 'charge_cards': charge_cards, 'discover_cards': discover_cards,
             'rush_cards': rush_cards, 'taunt_cards': taunt_cards}
for column in properties_type:
    df[column] = df[properties_type[column]].sum(axis=1)
for column in properties_rarity:
    df[column] = df[properties_rarity[column]].sum(axis=1)
    
df['common_and_class_cards'] = 30 - df['rare_cards'] - df['epic_cards'] - df['legendary_cards']
```

Давайте посмотрим, как менялась структура колод относительно редкости входящих в нее карт:

```python
def show_rarity_structure(Class, window, show_sets_flag = False, all_classes = False):  # Окно будет соответствовать дням, а не колодам!
    columns_rarity = ['common_and_class_cards', 'rare_cards', 'epic_cards', 'legendary_cards']

    if all_classes:
        grouped = df[columns_rarity].groupby(df['creation_date']).mean().rolling(window=window).mean()
        grouped = grouped.iloc[window:]
        fig = px.area(grouped, labels = {'value': 'Rarity structure', 'index': 'Creation date'}, title = 'Averaged rarity structure of all decks')
    else:
        df_class = df[df['Class'] == Class] 
        grouped = df_class[columns_rarity].groupby(df_class['creation_date']).mean().rolling(window=window).mean()
        grouped = grouped.iloc[window:]
        fig = px.area(grouped, labels = {'value': 'Rarity structure', 'index': 'Creation date'}, title = f'Rarity structure of {Class} decks')

    if show_sets_flag:
        for adventure in adventures:
            if adventures[adventure] < last_date_observed:
                fig.add_vline(x=adventures[adventure], line_width=2, line_color = 'blue', opacity=0.4)
        for expansion in expansions:
            if expansions[expansion] < last_date_observed:
                fig.add_vline(x=expansions[expansion], line_width=2, line_color = 'red', opacity=0.4)
    fig.show()
```

```python
show_rarity_structure('Hunter', 50, show_sets_flag = True)
```

```python
show_rarity_structure(None, 50, all_classes=True)
```

Можно также посмотреть на то, как менялась от патча к патчу структура колод у разных классов с точки зрения типа карт:

```python
def show_deck_structure(Class, window, show_sets_flag):
    columns_struct_1 = ['minion_count', 'spell_count', 'weapon_count'] 
    columns_struct_2 = ['minion_count', 'spell_count'] 

    df_class = df[df['Class'] == Class]

    if Class in ['Mage', 'Priest', 'Warlock', 'Druid']:  # да-да, есть у друида оружие, даже у чернокнижника есть, но здесь нет)
        grouped = df_class[columns_struct_2].groupby(df_class['creation_date']).mean().rolling(window=window).mean()
    else:
        grouped = df_class[columns_struct_1].groupby(df_class['creation_date']).mean().rolling(window=window).mean()
    grouped = grouped.iloc[window:]

    fig = px.area(grouped, labels = {'value': f'{Class} Deck Structure', 'index': 'Deck batches'})

    if show_sets_flag:
        for adventure in adventures:
            if adventures[adventure] < last_date_observed:
                fig.add_vline(x=adventures[adventure], line_width=2, line_color = 'blue', opacity=0.4)
        for expansion in expansions:
            if expansions[expansion] < last_date_observed:
                fig.add_vline(x=expansions[expansion], line_width=2, line_color = 'red', opacity=0.4)

    fig.show()
```

Наконец, давайте посмотрим на популярность различных кейвордов. Возьмем самые популярные: боевой клич, предстмертный хрип и таунт, и посчитаем отношение числа соответствующих карт к значению minion_count

```python
for column in properties_type:
    df[column + '_share'] = df[column]/df['minion_count']

properties_type_share = ['battlecry_cards_share' ,
             'deathrattle_cards_share' , 'stealth_cards_share' , 'charge_cards_share' , 'discover_cards_share',
             'rush_cards_share' , 'taunt_cards_share']


def show_keyword_popularity(keyword_list, window):
    modified_keyword_list = [key + '_cards_share' for key in keyword_list]
    keyword_share_grouped = df[properties_type_share].groupby(df['creation_date']).mean().rolling(window=window).mean()
    keyword_share_grouped = keyword_share_grouped.iloc[window:]
    keyword_share_grouped = keyword_share_grouped.reset_index()

    fig = px.line(keyword_share_grouped, x='creation_date', y=modified_keyword_list)

    for adventure in adventures:
        if adventures[adventure] < last_date_observed:
            fig.add_vline(x=adventures[adventure], line_width=2, line_color = 'blue', opacity=0.4)
    for expansion in expansions:
        if expansions[expansion] < last_date_observed:
            fig.add_vline(x=expansions[expansion], line_width=2, line_color = 'red', opacity=0.4)
                
    fig.show()

```

```python
show_keyword_popularity(['battlecry', 'deathrattle', 'taunt'], 10)
```

##### Посмотрим, наконец, на популярность самих карт! Интерактивная версия представлена в streamlit приложении)

```python
# Всякого рода подготовка
Classes = ['Mage', 'Priest', 'Warlock', 'Druid', 'Paladin', 'Hunter', 'Warrior', 'Rouge', 'Shaman']
uppercase_Classes = [s.upper() for s in Classes]

def get_api_data():
    with open('allcards.json', mode = 'r', encoding = 'utf-8') as file:
        data = json.load(file)
        return data
    
def get_popularity(card):
    
    def get_card_class(card_name, cards_data):
        for card in cards_data:
            if card.get('name') == card_name:
                return card.get('cardClass')
            
    card_class = get_card_class(card, get_api_data())
    
    if card_class in uppercase_Classes:
        df_classed = df[df['Class'] == card_class.capitalize()]
        card_data = df_classed[[card, 'creation_date']]
        

        popularity = card_data.groupby(card_data['creation_date']).mean().rolling(window=20).mean()
        popularity = popularity.reset_index()

    else:
        card_data = df[[card, 'creation_date']]
        popularity = card_data.groupby(card_data['creation_date']).mean().rolling(window=20).mean()
        popularity = popularity.reset_index()


    return popularity

def show_popularity(card):
    popularity = get_popularity(card)
    fig = px.line(popularity, x='creation_date', y=card, title=f'{card} popularity over time')
    
    for adventure in adventures:
        if adventures[adventure] < last_date_observed:
            fig.add_vline(x=adventures[adventure], line_width=2, line_color = 'blue', opacity=0.4)
    for expansion in expansions:
        if expansions[expansion] < last_date_observed:
            fig.add_vline(x=expansions[expansion], line_width=2, line_color = 'red', opacity=0.4)
            
    fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Average number in decks')
    fig.show()
```
